```python
#exista o legatura intre retail_price si cantitatea vanduta si suma incasata?#
#is there a link between retail_price, quantity sold and income?
df["price_dif"]=df["retail_price"]-df["price"]
df.head()
```

```python
def price(price):
    if price<= 0:
        return "nesemnificativa"
    elif price<10:
        return "semnificativa"
    else:
        return "foarte semnificativa"
    
```

```python
df["retail_price_influence"]=df["price_dif"].apply(price)
df["retail_price_influence"]
```

```python
plot_price=sns.lmplot(data=df, x="price_dif", y="units_sold", hue="retail_price_influence", palette="viridis", height=6, aspect=1.5)
plt.title("Retail Price Influence")
plt.show()
```

```python
df.pivot_table(index="rating", columns="merchant_rating",values="merchant_title", aggfunc="count")
```

```python
def rating(rating):
    if rating<= 1:
        return "one star"
    elif rating<=2:
        return "2 stars"
    elif rating<=3:
        return "3 stars"
    elif rating<=4:
        return "4 stars"
    else:
        return "5 stars"
```

```python
df["rating_m"]=df["merchant_rating"].apply(rating)
df
```

```python
df.groupby("rating_m").agg({"merchant_id":"count",
                           "units_sold":"sum",
                           "TotalSum_product":"sum"})
```

```python
df.groupby("rating_m").agg({"merchant_id":"count"}).plot.barh()
plt.show()
```

```python
df["rating_m"]=df["merchant_rating"].round(0).astype(int)
df["rating_m"]
```
